```python
##drop outliers for each column

#POSTS_CREATED
outlier_index = data_4_week[data_4_week['POSTS_CREATED']>200].index
data_4_week.drop(outlier_index, inplace= True)

#THANKYOUS_RECEIVED
outlier_index = data_4_week[data_4_week['THANKYOUS_RECEIVED']>90].index
data_4_week.drop(outlier_index, inplace= True)

#EVENTS_CREATED
outlier_index = data_4_week[data_4_week['EVENTS_CREATED']>15].index
data_4_week.drop(outlier_index, inplace= True)

#EVENT_PARTICIPANTS
outlier_index = data_4_week[data_4_week['EVENT_PARTICIPANTS']>30].index
data_4_week.drop(outlier_index, inplace= True)

#ITEMS_GIFTED
outlier_index = data_4_week[data_4_week['ITEMS_GIFTED']>20].index
data_4_week.drop(outlier_index, inplace= True)

#PLACES_RECOMMENDED
outlier_index = data_4_week[data_4_week['PLACES_RECOMMENDED']>40].index
data_4_week.drop(outlier_index, inplace= True)
```

```python
#visualzation
# Calculate summary statistics
df_4_clean = data_4_week[data_4_week['THANKYOUS_RECEIVED'].notnull()]['THANKYOUS_RECEIVED']


summary = data_4_week[data_4_week['THANKYOUS_RECEIVED'].notnull()]['THANKYOUS_RECEIVED'].describe()
mean = summary['mean']
median = summary['50%']
q1 = summary['25%']
q3 = summary['75%']

# Create a figure
plt.figure(figsize=(14, 7))

# Box Plot
sns.boxplot(data=df_4_clean, orient='h')

# Add annotations for Q1, median, Q3, and mean
plt.axvline(mean, color='g', linestyle='-', label=f'Mean: {mean}')
plt.axvline(median, color='b', linestyle='--', label=f'Median (Q2): {median}')
plt.axvline(q1, color='orange', linestyle=':', label=f'Q1: {q1}')
plt.axvline(q3, color='r', linestyle='-.', label=f'Q3: {q3}')

# Adding text for each annotation
plt.text(mean, 0.1, f'Mean: {mean:.2f}', color='g', ha='center')
plt.text(median, 0.2, f'Median: {median:.2f}', color='b', ha='center')
plt.text(q1, 0.3, f'Q1: {q1:.2f}', color='orange', ha='center')
plt.text(q3, 0.4, f'Q3: {q3:.2f}', color='r', ha='center')

# plt.title('Box Plot of THANKYOUS_RECEIVED')
plt.xlabel('THANKYOUS RECEIVED')

plt.show()
```
